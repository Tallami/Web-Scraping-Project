```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/processed/processed_prosettings.csv')
```

```python
top_countries = df['country'].value_counts().head(10)

sns.set(style='darkgrid')

plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='Set2')
plt.title('Top 10 Countries', fontsize=14, fontweight='bold')
plt.xlabel('Players')
plt.ylabel('Country')
plt.show()
```

```python
top_mouses = df['mouse'].value_counts().head(10)

sns.set(style='darkgrid')

plt.figure(figsize=(10, 6))
sns.barplot(x=top_mouses.values, y=top_mouses.index, palette='Set2')
plt.title('Top 10 Mouses', fontsize=14, fontweight='bold')
plt.xlabel('Players')
plt.ylabel('Mouse')
plt.show()
```

```python
top_dpi = df['dpi'].value_counts().head(3)

sns.set(style='darkgrid')

plt.figure(figsize=(10, 6))
sns.barplot(x=top_dpi.index, y=top_dpi.values, palette='Set2')
plt.title('Top 3 DPI', fontsize=14, fontweight='bold')
plt.xlabel('DPI')
plt.ylabel('Players')
plt.show()
```
